```python
# defines the encryption method to be used when the user correctly guesses the word
def decrypt(ciphertext):
    plaintext = ""
    i = 0
    
    while (i < len(ciphertext)):
        if (i % 2 == 0):
            cipher_ascii = ord(ciphertext[i])
            plain_ascii = cipher_ascii - 1
            
        else:
            cipher_ascii = ord(ciphertext[i])
            plain_ascii = cipher_ascii + 1
            
        plaintext += chr(plain_ascii)
        i += 1
        
    return plaintext
```

```python
import random # import the random module to enable random choice of words

words = ["WHDSPQZ", "XHO", "TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"] 
livesCount = 10 # number of lives
chosen = random.choice(words) # randomly chooses one of the encrypted words
chosen = str(chosen)
lettersUsed = "" # keeps track of the letters that have been used
currentWord = [] # stores the correct guesses and inital hyphens
chosenLength = len(chosen)
i = 0

# fills the current word being displayed with hyphens based on the chosen word's length
while (i < chosenLength):
    currentWord.append("-")
    i += 1

print(f"You have {str(livesCount)} lives left and have used these letters: {lettersUsed}")
print(f"Current Word: " + ' '.join(currentWord))
userGuess = input("Guess a letter: ")
userGuess = userGuess.upper() # makes the user's guess uppercase to allow for consistent comparison
userGuess = str(userGuess)

while (True):
    x = 0
    
    # replaces the hyphens with the letters in the correct position if the guess was correct
    for c in chosen:
            if (c == userGuess):
                currentWord[x] = c
                
            x += 1
    
    # check to see if the word has been correctly guessed/finished
    if ("-" not in currentWord):
        print(f"YAY! You guessed the word {chosen} !!")
        print("This was the encrypted message!")
        print("The decoded message says: " + str(decrypt(chosen)))
        break
    
    # check to see if the letter guessed was already guessed
    elif (userGuess in lettersUsed):
        print("You have already used that letter. Guess another letter.")
        print(f"You have {str(livesCount)} lives left and have used these letters: {lettersUsed}")
        print(f"Current Word: " + ' '.join(currentWord))
        userGuess = input("Guess a letter: ")
        userGuess = userGuess.upper()
        userGuess = str(userGuess)
    
    # check to see if the letter guessed was in the chosen word
    elif (userGuess in chosen):
        
        lettersUsed = lettersUsed + " " + userGuess
        print(f"You have {str(livesCount)} lives left and have used these letters: {lettersUsed}")
        print(f"Current Word: " + ' '.join(currentWord))
        userGuess = input("Guess a letter: ")
        userGuess = userGuess.upper()
        userGuess = str(userGuess)
    
    # else the guess was incorrect, and the number of lives should decrease by one
    else: 
            livesCount -= 1
            
            # if the number of lives has reached 0, the user has died
            if (livesCount == 0):
                print (f"Your letter, {userGuess} is not in the word.")
                print(f"You died, sorry. The word was {chosen}")
                print("You'll have to guess the word correctly to decode the message")
                break
            
            # otherwise, the user can continue guessing with one less life
            else:
                lettersUsed = lettersUsed + " " + userGuess
        
                print (f"Your letter, {userGuess} is not in the word.")
                print(f"You have {str(livesCount)} lives left and have used these letters: {lettersUsed}")
                print(f"Current Word: " + ' '.join(currentWord))
                userGuess = input("Guess a letter: ")
                userGuess = userGuess.upper()
                userGuess = str(userGuess)
        
```
